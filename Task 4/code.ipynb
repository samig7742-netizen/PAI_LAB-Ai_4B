```python
def check_placement_safety(grid, current_row, current_col, size):
    for i in range(current_row):
        if grid[i][current_col] == 1:
            return False
    r, c = current_row, current_col
    while r >= 0 and c >= 0:
        if grid[r][c] == 1:
            return False
        r -= 1
        c -= 1
    r, c = current_row, current_col
    while r >= 0 and c < size:
        if grid[r][c] == 1:
            return False
        r -= 1
        c += 1
    return True
def place_queens_recursive(grid, row_idx, size):
    if row_idx == size:
        return True
    for col_idx in range(size):
        if check_placement_safety(grid, row_idx, col_idx, size):
            grid[row_idx][col_idx] = 1
            if place_queens_recursive(grid, row_idx + 1, size):
                return True
            grid[row_idx][col_idx] = 0
    return False
def display_grid(grid):
    for row in grid:
        print(" ".join(map(str, row)))
board_dim = 4  
puzzle_grid = [[0 for _ in range(board_dim)] for _ in range(board_dim)]
if place_queens_recursive(puzzle_grid, 0, board_dim):
    print(f"Found a valid {board_dim}x{board_dim} solution:")
    display_grid(puzzle_grid)
else:
    print("No valid arrangement discovered.")
```
