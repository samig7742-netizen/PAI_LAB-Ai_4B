```python
def find_jug_solution(vol_a, vol_b, goal, history, steps):
    if (vol_a, vol_b) in history:
        return False
    history.add((vol_a, vol_b))
    steps.append((vol_a, vol_b))
    print(f"-> Step State: [A: {vol_a}L, B: {vol_b}L]")
    if vol_a == goal or vol_b == goal:
        print("\nSuccess: Goal volume reached!")
        return True
    possible_moves = [
        (4, vol_b),        
        (vol_a, 3),        
        (0, vol_b),        
        (vol_a, 0),        
        (vol_a - min(vol_a, 3-vol_b), vol_b + min(vol_a, 3-vol_b)),  
        (vol_a + min(vol_b, 4-vol_a), vol_b - min(vol_b, 4-vol_a))   
    ]
    for next_state in possible_moves:
        if find_jug_solution(next_state[0], next_state[1], goal, history, steps):
            return True
    steps.pop()
    return False
tracker = set()
find_jug_solution(0, 0, 2, tracker, [])
```
