### Analysis Cell

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
canvas = np.full((400, 400, 3), 255, dtype=np.uint8)
cv2.circle(canvas, (200, 200), 80, (80, 200, 120), -1)
cv2.rectangle(canvas, (100, 100), (300, 300), (0, 215, 255), 4)
cv2.putText(canvas, "CV Demo", (130, 350), cv2.FONT_HERSHEY_SIMPLEX, 1, (50, 50, 50), 2)
monochrome = cv2.cvtColor(canvas, cv2.COLOR_BGR2GRAY)
smoothed = cv2.medianBlur(monochrome, 5)
boundary_map = cv2.Canny(smoothed, 50, 150)
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
axes[0].imshow(cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB))
axes[0].set_title("Synthetic Image")
axes[0].axis('off')
axes[1].imshow(smoothed, cmap='gray')
axes[1].set_title("Median Filtering")
axes[1].axis('off')
axes[2].imshow(boundary_map, cmap='magma')
axes[2].set_title("Edge Detection Map")
axes[2].axis('off')
plt.tight_layout()
plt.show()
```
