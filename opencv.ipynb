```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Correct path syntax with raw string
image = cv2.imread(r'C:\Users\sami\OneDrive\Desktop\project1\images.jpg')

# Check if image is loaded
if image is None:
    print("Error: Image not found. Check the path!")
else:
    # Convert BGR to RGB
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    scale_factor_1 = 3.0  
    scale_factor_2 = 1/3.0
    height, width = image_rgb.shape[:2]

    # Zoom in
    new_height = int(height * scale_factor_1)
    new_width = int(width * scale_factor_1)

    zoomed_image = cv2.resize(src=image_rgb, 
                              dsize=(new_width, new_height), 
                              interpolation=cv2.INTER_CUBIC)

    # Scale down
    new_height1 = int(height * scale_factor_2)
    new_width1 = int(width * scale_factor_2)
    scaled_image = cv2.resize(src=image_rgb, 
                              dsize=(new_width1, new_height1), 
                              interpolation=cv2.INTER_AREA)

    # Plot images
    fig, axs = plt.subplots(1, 3, figsize=(12, 4))
    axs[0].imshow(image_rgb)
    axs[0].set_title('Original Image Shape:'+str(image_rgb.shape))
    axs[1].imshow(zoomed_image)
    axs[1].set_title('Zoomed Image Shape:'+str(zoomed_image.shape))
    axs[2].imshow(scaled_image)
    axs[2].set_title('Scaled Image Shape:'+str(scaled_image.shape))

    for ax in axs:
        ax.set_xticks([])
        ax.set_yticks([])

    plt.tight_layout()
    plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread(r'C:\Users\sami\OneDrive\Desktop\project1\images.jpg')
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
kernel = np.ones((3, 3), np.uint8)

dilated = cv2.dilate(image_gray, kernel, iterations=2)
eroded = cv2.erode(image_gray, kernel, iterations=2)
opening = cv2.morphologyEx(image_gray, cv2.MORPH_OPEN, kernel)
closing = cv2.morphologyEx(image_gray, cv2.MORPH_CLOSE, kernel)

fig, axs = plt.subplots(2, 2, figsize=(7, 7))
axs[0, 0].imshow(dilated, cmap='Greys'), axs[0, 0].set_title('Dilated Image')
axs[0, 1].imshow(eroded, cmap='Greys'), axs[0, 1].set_title('Eroded Image')
axs[1, 0].imshow(opening, cmap='Greys'), axs[1, 0].set_title('Opening')
axs[1, 1].imshow(closing, cmap='Greys'), axs[1, 1].set_title('Closing')

for ax in axs.flatten():
    ax.set_xticks([]), ax.set_yticks([])

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np

# Function to map each intensity level to output intensity level.
def pixelVal(pix, r1, s1, r2, s2):
    if (0 <= pix and pix <= r1):
        return (s1 / r1)*pix
    elif (r1 < pix and pix <= r2):
        return ((s2 - s1)/(r2 - r1)) * (pix - r1) + s1
    else:
        return ((255 - s2)/(255 - r2)) * (pix - r2) + s2

# Open the image.

image = cv2.imread(r'C:\Users\sami\OneDrive\Desktop\project1\images.jpg')
# Define parameters.
r1 = 70
s1 = 0
r2 = 140
s2 = 255

# Vectorize the function to apply it to each value in the Numpy array.
pixelVal_vec = np.vectorize(pixelVal)

# Apply contrast stretching.
contrast_stretched = pixelVal_vec(img, r1, s1, r2, s2)

# Save edited image.
cv2.imwrite('contrast_stretch.jpg', contrast_stretched)
```

```python
import cv2
import numpy as np

# Step 1: Load the image
image_path = r'C:\Users\sami\OneDrive\Desktop\project1\images.jpg'
image = cv2.imread(image_path)

# Step 2: Check if image loaded successfully
if image is None:
    print("Error: Image not found! Check the path.")
else:
    # Step 3: Get image dimensions
    height, width = image.shape[:2]
    quarter_height, quarter_width = height / 4, width / 4

    # Step 4: Translation matrix
    T = np.float32([[1, 0, quarter_width],
                    [0, 1, quarter_height]])

    # Step 5: Apply translation
    img_translation = cv2.warpAffine(image, T, (width, height))

    # Step 6: Display images
    cv2.imshow('Original Image', image)
    cv2.imshow('Translated Image', img_translation)

    cv2.waitKey(0)  # Wait until a key is pressed
    cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# Image path (local)
image_path = r'C:\Users\sami\OneDrive\Desktop\project1\Mountain.jpg'

# Read image
image = cv2.imread(image_path)

if image is None:
    print("Error: Image not found!")
else:
    pyramid = [image]

    # Create pyramid (downsampling)
    for i in range(3):
        image = cv2.pyrDown(image)
        pyramid.append(image)

    # Display pyramid levels (reverse order)
    for i in range(len(pyramid) - 1, -1, -1):
        print(f"Pyramid Level {i}")
        cv2.imshow(f'Pyramid Level {i}', pyramid[i])
        cv2.waitKey(0)

    cv2.destroyAllWindows()
```

```python
 
import cv2
import numpy as np

# Image path
image_path = r'C:\Users\sami\OneDrive\Desktop\project1\Mountain.jpg'

# Read image in grayscale (equalizeHist only works on grayscale)
img = cv2.imread(image_path, 0)

if img is None:
    print("Error: Image not found!")
else:
    # Histogram Equalization
    equ = cv2.equalizeHist(img)

    # Stack original and equalized images side by side
    res = np.hstack((img, equ))

    # Show result
    cv2.imshow('Original | Histogram Equalized', res)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
