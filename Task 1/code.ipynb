```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
```

### Analysis Cell

```python
raw_data = pd.read_csv("train.csv")
print(raw_data.head())
print(f"Dimensions: {raw_data.shape}")
```

### Analysis Cell

```python
print(raw_data.info())
print(raw_data.describe())
```

```python
null_counts = raw_data.isnull().sum()
print(null_counts[null_counts > 0])
```

### Analysis Cell

```python
selected_cols = [
    'LotArea',
    'YearBuilt',
    '1stFlrSF',
    '2ndFlrSF',
    'FullBath',
    'BedroomAbvGr',
    'TotRmsAbvGrd'
]
features_df = raw_data[selected_cols]
target_series = raw_data['SalePrice']
```

### Analysis Cell

```python
features_df = features_df.fillna(features_df.mean())
```

### Analysis Cell

```python
X_train, X_test, y_train, y_test = train_test_split(
    features_df, target_series, test_size=0.2, random_state=42
)
```

### Analysis Cell

```python
regressor = RandomForestRegressor(
    n_estimators=100,
    random_state=42
)
regressor.fit(X_train, y_train)
```

### Analysis Cell

```python
y_pred = regressor.predict(X_test)
error_metric = mean_absolute_error(y_test, y_pred)
print(f"Mean Absolute Error: {error_metric:.2f}")
```
