```python
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np
import json
print("Loading MiniLM model...")
model = SentenceTransformer('all-MiniLM-L6-v2')
data = [
    {"q": "How to stay calm?", "a": "Try deep breathing exercises and meditation."},
    {"q": "What is burn out?", "a": "Burnout is a state of emotional, physical, and mental exhaustion caused by excessive stress."}
]
texts = [item['q'] for item in data]
embeddings = model.encode(texts)
embeddings = np.array(embeddings).astype('float32')
index = faiss.IndexFlatL2(embeddings.shape[1])
index.add(embeddings)
print(f"Total vectors in index: {index.ntotal}")
query = "Tell me about feeling tired from work"
query_vec = model.encode([query]).astype('float32')
D, I = index.search(query_vec, 1)
print(f"Best Match Index: {I[0][0]}")
print(f"Answer: {data[I[0][0]]['a']}")
```
